```python
import numpy as np
import random
import csv
```

```python
def gen_gender():
    val = bool(random.getrandbits(1))
    if val: 
        return 'M'
    else:
        return 'F'
    
def gen_lat_long():
    latitude = 40.76
    longitude = -73.78
    dec_lat = random.random()/100
    dec_lon = random.random()/100
    return latitude + dec_lat, longitude + dec_lon

def gen_area():
#     50 % for Brooklyn & Manhattan,  50% for yhe others
    val0 = random.random()
    if val0 < 0.5: #50 % for Brooklyn & Manhattan
        val00 = random.random()
        if val00 < 0.5:
            return 'Brooklyn'
        else:
            return 'Manhattan'
    else:
        val01 = random.random()
        if val01 < 0.33:
            return 'Queens'
        elif val01 < 0.66:
            return 'Inwood'
        else:
            return 'Harlem'

def gen_room_type():
    val = bool(random.getrandbits(1))
    if val: 
        return 'Private room'
    else:
        return 'Entire room/apt'

def gen_minimum_nights(room_type):
    if room_type == 'Private room':
        return random.randint(1, 3)
    else:
        return random.randint(3, 5)

def gen_review_per_month(room_type):
    if room_type == 'Private room':
        return float("{0:.2f}".format(random.uniform(0, 15)))
    else:
        return float("{0:.2f}".format(random.uniform(0, 5)))

def gen_total_reviews(room_type):
    if room_type == 'Private room':
        val0 = random.random()
        if val0 < 0.33:
            return random.randint(1, 500)
        else:
            return random.randint(1, 250)
    else:
        return random.randint(1, 200)
    
def priceToLabel(price):
    if price < 100:
        return 'Low'
    elif price < 180:
        return 'Medium'
    else:
        return 'High'
```

```python
def gen_one_record():
    host_gender = gen_gender()
    lat, long = gen_lat_long()
    area = gen_area()
    room_type = gen_room_type()
    minimum_nights = gen_minimum_nights(room_type)
    review_per_month = gen_review_per_month(room_type)
    total_reviews = gen_total_reviews(room_type)
    
    price = random.randint(45, 55)
    if (area in {'Manhattan', 'Brooklyn'}):
        price = random.randint(65, 70) * 2
    if (room_type == 'Entire room/apt'):
        price *=2
    if (review_per_month > 5):
        price += 25
    if (host_gender == 'F'):
        price += 15
        
    label = priceToLabel(price) 
    return [host_gender, lat, long, area, room_type, minimum_nights, review_per_month, total_reviews, label]
```

```python

labels = []
for i in range(100000):
    labels.append(gen_one_record())

with open('data.csv', 'w', newline='') as myfile:
    wr = csv.writer(myfile)
    wr.writerow(['Host_Gender', 'Latitude', 'Longitude', 'Area', 'Room_type', 'Minimum_nights', 'review_per_month', 
                 'total_reviews', 'price'])
    wr.writerows(labels)
```
